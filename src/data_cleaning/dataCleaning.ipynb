# Download the data

```python
import os
import pandas as pd
import urllib.request

urls = ["http://capstone-music-similarity.s3-website-us-east-1.amazonaws.com/raw_albums.csv",
        #"http://capstone-music-similarity.s3-website-us-east-1.amazonaws.com/raw_artists.csv",
        #"http://capstone-music-similarity.s3-website-us-east-1.amazonaws.com/raw_genres.csv",
        "http://capstone-music-similarity.s3-website-us-east-1.amazonaws.com/raw_echonest.csv",
        "http://capstone-music-similarity.s3-website-us-east-1.amazonaws.com/raw_tracks.csv"]

print("Loading data from URLs...")
for url in urls:
    filename = url.split("/")[-1]
    print(f"Processing {filename}...")
    if not os.path.exists(f"../data_cleaning/raw_data/{filename}"):
        urllib.request.urlretrieve(url, f"../data_cleaning/raw_data/{filename}")
        print(f"{filename} : downloaded")
    else:
        print(f"{filename} : already exists, skipping download")
    
```

```python
raw_albums = pd.read_csv("../data_cleaning/raw_data/raw_albums.csv")
print(raw_albums.head(5))
print(raw_albums.info())
```
